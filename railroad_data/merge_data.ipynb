```python
import pandas as pd
```

```python
df_query = pd.read_csv("geocoded_querydata-16-5-2023 14_02_28.csv")
df_railroad = pd.read_csv("railroad_data.csv")
df_unemployment_data = pd.read_excel("unemployment_data.xlsx")
```

```python
df_unemployment_data.rename({"rate":"Unemployment_rate"},axis=1,inplace=True)
df_unemployment_data.drop("Rank",axis=1,inplace=True)
df_unemployment_data.head() 
```

```python
df_railroad.columns = [str(column)+ "_railroad" for column in df_railroad.columns]
df_railroad.head()
```

```python
df_query.rename({"state":"State"},axis=1,inplace=True)
df_query.head()
```

```python
df_merged = pd.merge(df_unemployment_data, df_railroad, left_on="State", right_on="State_railroad").drop("State_railroad",axis=1)
df_total = pd.merge(df_query,df_merged,left_on="State",right_on="State")
df_total.head()
```

```python
df_total.to_excel("Combined_dataset.xlsx")
df_total.shape
```
