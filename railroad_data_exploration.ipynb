```python
import pandas as pd
```

# Merge the railroad data

```python
data1 = pd.read_csv('Railroads Data\Railroad Retirement Payments by State.csv')
data2 = pd.read_csv('Railroads Data\Freight Rail Average Wages & Benefits Per Employee.csv')
data3 = pd.read_csv('Railroads Data\Freight Rail Employment.csv')
data4 = pd.read_csv('Railroads Data\Freight Railroads by State.csv')
data5 = pd.read_csv('Railroads Data\Originated Rail Carloads by State.csv')
data6 = pd.read_csv('Railroads Data\Originated Rail Tons by State.csv')
data7 = pd.read_csv('Railroads Data\Railroad Retirement Beneficiaries by State.csv')
data8 = pd.read_csv('Railroads Data\Terminated Rail Carloads by State.csv')
data9 = pd.read_csv('Railroads Data\Terminated Rail Tons by State.csv')
data10 = pd.read_csv('Railroads Data\Total Rail Miles by State.csv')

output1 = pd.merge(data1, data2[['State', 'Freight Rail Average Wages & Benefits Per Employee']],
				on='State',
				how='left')

output2 = pd.merge(output1, data3[['State', 'Freight Rail Employees']],
				on='State',
				how='left')

output3 = pd.merge(output2, data4[['State', 'Number of Railroads']],
				on='State',
				how='left')

output4 = pd.merge(output3, data5[['State', 'Originated Carloads']],
				on='State',
				how='left')

output5 = pd.merge(output4, data6[['State', 'Originated Tons (millions)']],
				on='State',
				how='left')

output6 = pd.merge(output5, data7[['State', 'Retirement Beneficiaries']],
				on='State',
				how='left')

output7 = pd.merge(output6, data8[['State', 'Terminated Carloads']],
				on='State',
				how='left')

output8 = pd.merge(output7, data9[['State', 'Terminated Tons (millions)']],
				on='State',
				how='left')

output9 = pd.merge(output8, data10[['State', 'Total Rail Miles']],
				on='State',
				how='left')

print(output3)
```

```python
output9.to_csv('data_merge.csv', index = False)  # Export merged pandas DataFrame
```

```python
merged_data = pd.read_csv('data_merge.csv')
merged_data
```

# Fix the railraod data

```python
# convert rows into float and remove $ and commas : Retirement Payments ($ mil), Freight Rail Average Wages & Benefits Per Employee 
merged_data['Retirement Payments ($ mil)'] = merged_data['Retirement Payments ($ mil)'].replace({'\$': '', ',': ''}, regex=True).astype(float)
merged_data['Freight Rail Average Wages & Benefits Per Employee'] = merged_data['Freight Rail Average Wages & Benefits Per Employee']\
                                                .replace({'\$': '', ',': ''}, regex=True).astype(float)

# convert rows into floats and remove commas: Freight Rail Employees, Originated Carloads, Retirement Beneficiaries, Terminated Carloads, Total Rail Miles

merged_data['Freight Rail Employees'] = merged_data['Freight Rail Employees'].replace({',': ''}, regex=True).astype(float)
merged_data['Originated Carloads'] = merged_data['Originated Carloads'].replace({',': ''}, regex=True).astype(float)
merged_data['Retirement Beneficiaries'] = merged_data['Retirement Beneficiaries'].replace({',': ''}, regex=True).astype(float)
merged_data['Terminated Carloads'] = merged_data['Terminated Carloads'].replace({',': ''}, regex=True).astype(float)
merged_data['Total Rail Miles'] = merged_data['Total Rail Miles'].replace({',': ''}, regex=True).astype(float)

merged_data
```

```python
# remove District of Columbia since it only has 2 values and the rest is NaN
merged_data = merged_data[merged_data['State'] != 'District of Columbia']

merged_data
```

```python
# Originated Tons (millions) value (a) converted to < 50000 (0.05 million)
merged_data.loc[merged_data['Originated Tons (millions)'] == '(a)', 'Originated Tons (millions)'] = '<0.05'

merged_data
```
